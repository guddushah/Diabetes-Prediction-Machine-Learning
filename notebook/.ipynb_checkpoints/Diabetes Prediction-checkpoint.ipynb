### Removing Outliers and applying Log-Normal Transformation on the Skewed data

```python
df=data[['Pregnancies', 'SkinThickness', 'Insulin', 'DiabetesPedigreeFunction', 'Age']]
for feature in df:
    df[feature] = np.where(df[feature] == 0, 0, np.log10(df[feature]))
```

```python
df.head()
```

### Merging data on the basis of columns

```python
df_merged = pd.concat([df, data[['Glucose','BloodPressure','BMI','Outcome']]], axis=1)
df_merged.head()
```

## Model Training

### Separating the Independent and Dependent features

```python
X = df_merged.drop('Outcome',axis=1)
X.head()
```

```python
y = df_merged.Outcome
y.head()
```

### Scaling the features using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
df=X.copy()
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(df),columns = df.columns)
```

```python
X_scaled.head()
```

### Train-Test Split of data

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=42)
X_train.shape, X_test.shape
```

### Creating Logistic Regression Model

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

### Accuracy of the model

```python
accuracy = accuracy_score(y_pred,y_test)
print(accuracy)
```

### Classification Report

```python
print(classification_report(y_test, y_pred))
```

### Confusion Matrix

```python
conf_mat = confusion_matrix(y_test, y_pred)
conf_mat
```

```python
import seaborn as sns
plt.figure(figsize=(8,6), dpi=100)
sns.set(font_scale = 1.1) 
ax = sns.heatmap(conf_mat, annot=True, fmt='d', )
ax.set_xlabel("Predicted", fontsize=14, labelpad=20)
ax.xaxis.set_ticklabels(['0', '1'])
ax.set_ylabel("Actual", fontsize=14, labelpad=20)
ax.yaxis.set_ticklabels(['0', '1'])
ax.set_title("Confusion Matrix for the Diabetes Prediction", fontsize=14, pad=20)
 
plt.show()
```
